```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
%matplotlib inline
```

```python
design_dataset=pd.read_csv('design.csv')
yarn_consumption_dataset=pd.read_csv('yarn-consumption.csv')
results_dataset=pd.read_csv('results.csv')
```

```python
design_dataset.head()
```

```python
yarn_consumption_dataset.head()
```

```python
results_dataset.head()
```

```python
new_data_1=pd.merge(design_dataset, yarn_consumption_dataset, on='DesignId')
```

```python
new_data_1.head()
```

```python
new_data_1.replace(0, np.nan, inplace= True)
new_data_1.replace(np.nan, np.round(new_data_1.mean(), decimals=1), inplace= True)
```

```python
# new_data_1.to_csv('new_data_1.csv', sep='\t')
```

```python
final_dataset=pd.merge(new_data_1, results_dataset, on=['DesignId','PRL','PRW','Stretch'])
```

```python
final_dataset.head()
```

```python
final_dataset.to_csv('final_dataset.csv', sep='\t')
```

```python
X=final_dataset.iloc[:,[4,5,2]]
y=final_dataset.iloc[:,[12,13,14]]
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression 
regressor =LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
new_PRL = float(input('Enter the PRL value :'))
new_PRW =float(input('Enter the PRW value :'))
new_stretch=float(input('Enter the stretch value :'))
```

```python
predict_values=regressor.predict([[new_PRL ,new_PRW,new_stretch]])
print('predict values FRJ:',np.round(predict_values[0][0],decimals=2),'\n')
print('predict values FRP:',np.round(predict_values[0][1],decimals=2),'\n')
print('predict values FRL:',np.round(predict_values[0][2],decimals=2),'\n')
```
